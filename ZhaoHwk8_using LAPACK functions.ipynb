### 2.a Make a (large) SPD matrix

```julia
function generateSPDmatrix(n)
    A = rand(n,n)
    B = Symmetric(A,:U)
    B = B + n*Matrix(I(n))
    return B
end
```

```julia
SPD = generateSPDmatrix(1000);
```

```julia
SPD
```

### 2.b Compare the times taken to find the eigenvectors of the matrix,

```julia
@timev vecs = eigvecs(SPD)
```

```julia
@timev evals = eigvals(SPD)
```

```julia
@timev W,VL,VR = LAPACK.geev!('V','V',SPD)
```

```julia
W
```

```julia
VR
```

```julia
VL
```

### Result: geev! takes almost the same time as eigvecs.  The eigenvaluse are same but in different order, thus the eigenvectors are not same. It seems that the signs of vectors of geev! are different with corresponding vectors of eigvecs.

### 3.a make a random (large) overdetermined system 

```julia
A_over = rand(200,100);
b_over = rand(200);
```

### 3.b  compare the times

```julia
@timev x_1 = A_over \ b_over
```

```julia
@timev F,B,ssr = x_2 = LAPACK.gels!('N',A_over,b_over)
```

```julia
B
```

### Result: the gels! function of LAPACK is faster, which speedup about 10 times than the \ operator

### 4.a make a (large) random square matrix
